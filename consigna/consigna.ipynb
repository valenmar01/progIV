# Facultad de Ciencias de la Administración - UNER
## TRABAJO FINAL INTEGRADOR
### Programación IV – 2026 1er cuatrimestre
### Licenciatura en Sistemas

**OBJETIVOS**
El Trabajo Final Integrador tiene como objetivos que el estudiante:
* Defina la estructura de los documentos web y las relaciones entre estos.
* Establezca estilos apropiados para dispositivos de diferentes resoluciones.
* Diseñe e implemente servicios web basados en arquitecturas REST.
* Utilice estructuras de datos apropiadas para la gestión eficiente de la información.

**CONDICIONES DE ENTREGA**
El Trabajo Final Integrador deberá ser:
* Realizado en forma grupal.
* Los grupos deberán contar con un mínimo de 3 y un máximo de 5 alumnos.
* Cargado en la sección del Campus Virtual como un archivo de texto plano (txt) que cuente con la URL al repositorio GitHub. Además, para asegurar el acceso del equipo docente, se sugiere cargar un archivo comprimido en formato ZIP, RAR, TAR.GZ u otro formato similar con el proyecto completo excluyendo las dependencias (node_modules).
* Indicar número de grupo, apellido y nombre de los integrantes.
* Entregado antes de la fecha límite informada en el campus.
* Las soluciones deben ser de autoría propia. Aquellas que se detecten como idénticas entre diferentes grupos serán desaprobadas para todos los involucrados.
* Se valorarán la exactitud, eficiencia y prolijidad (identación y otras buenas prácticas) de las soluciones planteadas, así como también la calidad de la exposición realizada.
* Los integrantes del equipo deberán realizar una exposición del proyecto, detallando su funcionamiento y las decisiones de diseño tomadas.

**Requisitos Funcionales**
Es necesario contar con una aplicación web disponible en la intranet de la institución que le permita al personal afectado llevar a cabo las siguientes tareas:
* Login de usuario o pantalla de inicio de sesión.
* Panel de resumen o dashboard (Totales de cursos, estudiantes y links rápidos a cursos activos).

Respecto de los estudiantes:
* BREAD completo.
* Incluir distintos criterios de búsqueda.
* Tener en cuenta paginación de resultados.

En referencia a cursos:
* BREAD completo.
* Incluir distintos criterios de búsqueda.
* Tener en cuenta paginación de resultados.
* Generación / Impresión de Diplomas.

En relación a Inscripciones:
* BREAD exceptuando NO aplica edición.
* Tener en cuenta paginación de resultados.
* Incluir controles que no permitan la inscripción adicional de estudiantes si se supera el cupo establecido por la columna inscriptos_max o la inscripción no está habilitada.
* No permitir inscripciones duplicadas.
* Generación / Impresión de diploma individual.

**Funcionalidades requeridas y requerimientos técnicos**
* Aplicación con interfaz web utilizando diseño responsivo.
* Autenticación de usuarios utilizando tokens JWT.
* API Rest cumpliendo con las buenas prácticas de diseño de APIs Rest.
* Generación de Certificados en PDF.
* Gestionar información sensible del servicio mediante variables de entorno.
* No utilizar borrados físicos sino "soft deletes", es decir utilizar los campos de estado pensados para indicar que los registros no están activos.

**Opcionales:**
* Envío de email de confirmación de inscripción.
* Pantalla de cambio de contraseña de usuario.
* Código QR en Diploma para verificar autenticidad.
* BREAD de usuarios.

**Notas:**
BREAD es el acrónimo de:
* Browse: Mostrar todos los recursos.
* Read: Leer / Mostrar un único recurso.
* Edit: Actualizar / Modificar un recurso.
* Add: Agregar / Crear un nuevo recurso.
* Delete: Eliminar / Borrar un recurso.
